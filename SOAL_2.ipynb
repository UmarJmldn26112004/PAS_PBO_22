```python
#untuk menampilkan grafik scatter
import matplotlib.pyplot as plt

df.plot(kind='scatter',x='Life expectancy', y='percentage expenditure')

plt.title('Grafik scatter')
plt.xlabel('Life expectancy')
plt.ylabel('percentage expenditure')
plt.xlim(0,80)
plt.ylim(0,10)
plt.grid(True)
plt.show()
```

```python
import numpy as np
X = np.array(df['Life expectancy'])
y = np.array(df['percentage expenditure'])

print(f'X : {X}')
print(f'y : {y}')
```

```python
X = X.reshape(-1, 1)

X.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python
X_vis = np.array([0, 80]).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis,y_vis, '-r')
plt.title('Grafik')
plt.xlabel('Life expectancy')
plt.ylabel('percentage expenditure')
plt.xlim(0,80)
plt.ylim(0,10)
plt.grid(True)
plt.show()
```
